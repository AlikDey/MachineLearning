```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("C:\\Users\\2151491\\Downloads\\MachineLearningDocs\\MonroeTwn.xlsx")
df
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.Area,df.Price,color='red', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.Area,df.Price,color='red', marker='*')
plt.plot(df.Area,reg.predict(df[['Area']]),color='blue')
```

```python
dp = pd.read_csv("C:\\Users\\2151491\\Downloads\\MachineLearningDocs\\PriceToPredict.csv")
dp.head(5)
```

```python
dp['Price'] = reg.predict(dp)
```
