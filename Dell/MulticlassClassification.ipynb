```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[71])
```

```python
digits.target[71]
```

```python
model.predict([digits.data[71]])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
fig, ax = plt.subplots()
cax = plt.imshow(cm, cmap=plt.cm.Blues, alpha=0.3, interpolation='nearest')
fig.colorbar(cax)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        if(i==j):
            ax.text(j,i,str(cm[i,j]),ha='center',va='center',color='white')
        else:
            ax.text(j,i,str(cm[i,j]),ha='center',va='center',color='lightgray')
plt.xlabel('Predicetd')
plt.ylabel('Truth')
```

```python
ax = sn.heatmap(cm, annot=True)
ax.set_xlabel('Predicted')
ax.set_ylabel('Truth') 
```
