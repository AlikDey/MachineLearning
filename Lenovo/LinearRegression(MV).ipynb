```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("A:\\Machine Learning\\hiring.csv")
df
```

```python
df.experience = df.experience.fillna('zero')
df
```

```python
from word2number import w2n
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
import math
test_score_median = math.floor(df['test_score(out of 10)'].median())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(test_score_median)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
dp = pd.read_csv("A:\\Machine Learning\\HiringToPredict.csv")
dp
```

```python
dp['price($)'] = reg.predict(dp)
dp
```

```python
dp.to_csv("A:\\Machine Learning\\HiringPredicted.csv",index=False)
```
