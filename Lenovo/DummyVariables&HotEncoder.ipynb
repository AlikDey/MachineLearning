Demmy Variable

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("A:\\Machine Learning\\carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
df_res = pd.concat([df, dummies], axis='columns')
df_res
```

```python
fdf = df_res.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
fdf
```

```python
X = fdf.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
model = linear_model.LinearRegression()
```

```python
y = fdf['Sell Price($)']
y
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
print("Mercedez with 4yr age and having 45000 milage: ",model.predict([[45000,4,0,0]]),"\nBMW with 7yr age and having 86000 milage: ",model.predict([[86000,7,0,1]]))
```

OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfen = df
dfen['Car Model'] = le.fit_transform(dfen['Car Model'])
dfen
```

```python
x = dfen[['Car Model','Mileage','Age(yrs)']].values
x
```

```python
Y = dfen['Sell Price($)']
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe = OneHotEncoder()
column_trans = ColumnTransformer(transformers=[('cat', ohe, [0])],remainder='passthrough')
```

```python
x = column_trans.fit_transform(x)
x
```

```python
x = x[:,1:]
x
```

```python
model_ohe = linear_model.LinearRegression()
model_ohe.fit(x,Y)
```

```python
model_ohe.score(x,Y)
```

```python
print("Mercedez with 4yr age and having 45000 milage: ",model_ohe.predict([[0,1,45000,4]]),"\nBMW with 7yr age and having 86000 milage: ",model_ohe.predict([[1,0,86000,7]]))
```
